```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Technique to handle outliers
# 1. Remove Outliers
# 2. Floor the Outliers (Replace extreme outliers with the maximum or minimum value)
# 3. Imputation (Replace with mean, median, mode)
# If we know the bounded value and don't have outlier, use min max normalization else use z-score
```

```python
# Replace with Q1 if value is below boundary and if value is above boundary replace with Q3
df['Age'] = df['Age'].apply(lambda x: Q1 if x < lower_bound else (Q3 if x > upper_bound else x))
```

# **Building a Logistic Regression Model**

A guide to creating and fitting a logistic regression model for binary classification.

```python
# Split the data into dependent and independent features
independent_data = df_encoded.drop('Survived', axis= ? )
dependent_data = df_encoded['Survived']
```

```python
independent_data
```

# **Split the Data into Train and Test**

Dividing the dataset into training and testing subsets to evaluate model performance.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(independent_data, dependent_data, test_size=0.2, random_state=101)
# random_state control the randomness of the data splitting process
# get same data for training and testing every time we run the code
```

```python
X_train.shape
```

# **Performing Prediction Using Logistic Regression**

Using the logistic regression model to make predictions on the test data.

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
predicitions = logmodel.predict(X_test)
```

```python
predicitions
```

```python
print(y_test.tolist())
```

# **Analyzing the Confusion Matrix**

Evaluating the model's performance by examining the confusion matrix, which shows true positives, false positives, true negatives, and false negatives.

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, predicitions)
cm
```

```python
# plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Survived', 'Survived'],
            yticklabels=['Not Survived', 'Survived'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

# **Calculating Model's Accuracy**

Calculating the percentage of correct predictions made by the model.

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy = accuracy_score(y_test, predicitions)
accuracy
```

# **Displaying Which Feature Has More Contribution to the Output**

Analyzing which input features have the most influence on the model's predictions.

```python
coefficients = logmodel.coef_[0]
coefficients
```

```python
coef_df = pd.DataFrame({'Feature': X_train.columns, 'Coefficient': coefficients})
coef_df
```

```python
coef_df['Abs_Coefficient'] = coef_df['Coefficient'].abs()
# (ignoring the sign, since it indicates the direction of the relationship)
```

- **PassengerId**: Unique ID assigned to each passenger in the dataset.
- **Survived**: Survival status of the passenger (0 = Did not survive, 1 = Survived).
- **Pclass**: Passenger class (1 = 1st class, 2 = 2nd class, 3 = 3rd class).
- **Name**: Name of the passenger.
- **Sex**: Gender of the passenger (male or female).
- **Age**: Age of the passenger in years.
- **SibSp**: Number of siblings or spouses aboard the Titanic with the passenger.
  - **Sib**: Siblings (brother, sister, stepbrother, stepsister).
  - **Sp**: Spouse (husband or wife, not including fiancé).
- **Parch**: Number of parents or children aboard the Titanic with the passenger.
  - **Par**: Parents (father, mother, stepfather, stepmother).
  - **Ch**: Children (son, daughter, stepson, stepdaughter).
- **Ticket**: Ticket number assigned to the passenger.
- **Fare**: Fare paid for the ticket (in British pounds).
- **Cabin**: Cabin number where the passenger was accommodated (if available).
- **Embarked**: Port of embarkation where the passenger boarded the Titanic:
  - **C**: Cherbourg.
  - **Q**: Queenstown.
  - **S**: Southampton.

```python
coef_df
```

```python
sns.barplot(x="Feature", y="Abs_Coefficient", data=coef_df, palette="viridis")
plt.xlabel("Features", fontsize=12)
plt.ylabel("Absolute Coefficient", fontsize=12)
plt.title("Feature Importance (Absolute Coefficients)", fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```
